```python
book_recommemd_data_table=book_recommemd_data1.pivot_table(index='User_ID',columns='Book_Title',values='Book_Rating').reset_index(drop=True)
```

```python
book_recommemd_data_table
```

```python
book_recommemd_data_table.index=book_recommemd_data1.User_ID.unique()
```

```python
book_recommemd_data_table.fillna(0,inplace=True)
```

```python
book_recommemd_data_table
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
book_recommemd_user_similarity=1-pairwise_distances(book_recommemd_data_table.values,metric='cosine')
```

```python
book_recommemd_user_similarity
```

```python
#store results in dataframe
book_recommemd_user_similarity_df=pd.DataFrame(book_recommemd_user_similarity)
```
